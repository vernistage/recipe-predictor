```python
def process_line_item(dir_list):
    try:
        directions = ",".join(dir_list)
        gensim_dir = gensim.utils.simple_preprocess(directions)
        return gensim_dir
    except:
        return ''
```

```python
df["gensim_directions"] = df["directions"].map(process_line_item)
```
