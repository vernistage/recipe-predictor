```python
def process_line_item(dir_list):
    try:
        directions = ",".join(dir_list)
        gensim_dir = gensim.utils.simple_preprocess(directions)
        return gensim_dir
    except:
        return ''
```

```python
df["gensim_directions"] = df["directions"].map(process_line_item)
```

```python
df.head()
```

```python
# build vocabulary and train model
model = gensim.models.Word2Vec(
    df["gensim_directions"],
    size=150,
    window=10,
    min_count=2,
    workers=10)
model.train(df["gensim_directions"], total_examples=len(df["gensim_directions"]), epochs=10)
```

```python
# create pickle
import pickle
modelname = "../word2vec_model"
file_object = open(modelname,'wb')
pickle.dump(model, file_object)
```

```python
# use pickle
with open(modelname, 'rb') as f:
    word2vec_pickle_model = pickle.load(f)   
```

```python
download('punkt') #tokenizer, run once
download('stopwords') #stopwords dictionary, run once
stop_words = stopwords.words('english')

def preprocess(text):
    text = text.lower()
    doc = word_tokenize(text)
    doc = [word for word in doc if word not in stop_words]
    doc = [word for word in doc if word.isalpha()] #restricts string to alphabetic characters only
    return doc
```

```python
def list_to_text(my_list):
    try:
        my_string = " ".join(my_list)
        return my_string
    except:
        return ''
```

```python
from gensim.models.doc2vec import Doc2Vec
df["gensim_directions_doc"] = df["gensim_directions"].map(list_to_text)
```
